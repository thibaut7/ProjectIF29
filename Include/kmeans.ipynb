### Nous allons normaliser les donnees pour avoir des donnees plus homogenes

```python
X = StandardScaler().fit_transform(X)
```

```python
clustering = KMeans(n_clusters=2)
clustering.fit(X)
```

```python
labels = clustering.labels_
```
